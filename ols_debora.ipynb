```python
import numpy as np
import matplotlib.pyplot as plt

# Dados
data = {
    "Horas Estudadas": [4.4, 9.6, 7.6, 6.4, 2.4, 2.4, 1.5, 8.8, 6.4, 7.4, 1.2, 9.7, 8.5, 2.9, 2.6],
    "Nota na Prova": [23.33, 48.12, 37.71, 31.34, 9.06, 10.58, 6.69, 46.09, 32.74, 33.34, 6.57, 47.88, 41.11, 15.78, 15.24]
}

# Convertendo para arrays numpy
horas_estudadas = np.array(data["Horas Estudadas"])
notas_prova = np.array(data["Nota na Prova"])

# Cálculos manuais de b0 e b1
mean_x = np.mean(horas_estudadas)
mean_y = np.mean(notas_prova)
b1 = np.sum((horas_estudadas - mean_x) * (notas_prova - mean_y)) / np.sum((horas_estudadas - mean_x)**2)
b0 = mean_y - b1 * mean_x

# b
y_pred = b0 + b1 * horas_estudadas

# SSE - Soma dos Quadrados dos Erros
sse = np.sum((notas_prova - y_pred)**2)

# Coeficiente de determinação (R^2)
ss_total = np.sum((notas_prova - mean_y)**2)
r_squared = 1 - (sse / ss_total)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(horas_estudadas, notas_prova, color='gray', label='Dados Observados')
plt.plot(horas_estudadas, y_pred, color='red', label=f'Regressão Linear: y = {b0:.2f} + {b1:.2f}x\n$R^2$ = {r_squared:.2f}\nSSE = {sse:.2f}')
plt.xlabel('Horas Estudadas')
plt.ylabel('Nota na Prova')
plt.title('Diagrama de Dispersão: Horas Estudadas vs. Nota na Prova')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Dados
data = {
    "Horas Estudadas": [4.4, 9.6, 7.6, 6.4, 2.4, 2.4, 1.5, 8.8, 6.4, 7.4, 1.2, 9.7, 8.5, 2.9, 2.6],
    "Nota na Prova": [23.33, 48.12, 37.71, 31.34, 9.06, 10.58, 6.69, 46.09, 32.74, 33.34, 6.57, 47.88, 41.11, 15.78, 15.24]
}

# Convertendo para arrays numpy
horas_estudadas = np.array(data["Horas Estudadas"]).reshape(-1, 1)
notas_prova = np.array(data["Nota na Prova"])

# Regressão Linear com scikit-learn
model = LinearRegression()
model.fit(horas_estudadas, notas_prova)

# Previsões
y_pred = model.predict(horas_estudadas)

# Coeficientes
b0 = model.intercept_
b1 = model.coef_[0]

# SSE
sse = mean_squared_error(notas_prova, y_pred) * len(notas_prova)

# R^2
r_squared = r2_score(notas_prova, y_pred)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(data["Horas Estudadas"], data["Nota na Prova"], color='gray', label='Dados Observados')
plt.plot(data["Horas Estudadas"], y_pred, color='red', label=f'Regressão Linear: y = {b0:.2f} + {b1:.2f}x\n$R^2$ = {r_squared:.2f}\nSSE = {sse:.2f}')
plt.xlabel('Horas Estudadas')
plt.ylabel('Nota na Prova')
plt.title('Diagrama de Dispersão: Horas Estudadas vs. Nota na Prova')
plt.grid()
plt.legend()
plt.show()
```
